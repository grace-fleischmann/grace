```python
import numpy as np 
import matplotlib.pyplot as plt
def make_sine_wave(x, A, w): 
    return A * np.sin(w*x)
x = np.linspace(0, 2* np.pi, 1000)
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
figs, axes = plt.subplots(1, 3, figsize=(25, 5))
for A, w in zip(amplitude, frequency):
    y1 = make_sine_wave(x, A, 1)
    axes[0].scatter(x, y1, label = f"A={A}")
    axes[0].set_xlabel("x-axis")
    axes[0].set_ylabel("y-axis")
    y2 = make_sine_wave(x, 1, w)
    axes[1].scatter(x, y2, label = f"w={w}")
    axes[1].set_xlabel("x-axis")
    axes[1].set_ylabel("y-axis")
    y3 = make_sine_wave(x, A, w)
    axes[2].scatter(x,y3, label = f"A{A}, w = {w}")
    axes[2].set_xlabel("x-axis")
    axes[2].set_ylabel("y-axis")
axes[0].set_title("Varied Amplitude", fontsize=22)
axes[1].set_title("Varied Frequency", fontsize=22)
axes[2].set_title("Varied Amplitude and Frequency", fontsize=22)
axes[0].legend(fontsize = 14)
axes[1].legend(fontsize = 14)
axes[2].legend(fontsize = 14)
plt.show()
```
