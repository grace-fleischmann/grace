```python
arr = np.array([
    [2, 3, 5],
    [14, 6, 8],
    [11, 13, 17],
    [7, 10, 13]
])
def is_prime(n):
    if n<2:
        return False 
    for i in range(2,n):
        if n % i == 0:
            return False 
    return True 
def alien_star(array): 
    result = []
    for row in array:
        for num in row: 
                if is_prime(num):
                    result.append(row)
                    break
        return np.array(result)
    
print(alien_star(arr))
```

```python
arr2 = np.zeros((8,8))
for i in range(arr2.shape[0]):
    for j in range(arr2.shape[1]):
        if (i + j) %2 == 0:
            arr2[i,j] = 1
print(arr2)
def reverse_checkerboard(board):
    for j in range(board.shape[0]):
        for i in range(board.shape[1]):
            if (i + j) % 2 == 0:
                board[i,j] = 0
            else:
                board[i,j] = 1
    return board
                
print(reverse_checkerboard(arr2))
```

```python
def expand_universe(input_str, num):
    string_array = list(input_str)
    spaced_array = []
    for i in range(len(string_array)): 
        spaced_array.append(string_array[i])
        if i != len(string_array)-1:
            for j in range(num): 
                spaced_array.append(" ")
    return ''.join(spaced_array)
expand_universe("test", 2)
```

```python
star_array = ([[1123, 1456, 1789, 1324, 1876],
[1567, 1987, 1678, 1405, 1589],
[1345, 1654, 1523, 1109, 1923],
[1298, 1890, 1367, 1784, 1432],
[1823, 1756, 1489, 1672, 1550]])
def find_guide_star(array):
    result = []
    for row in array:
        sorted_row = np.sort(row)
        result.append(sorted_row[-2])
    return np.array(result)
print(find_guide_star(star_array))
```
